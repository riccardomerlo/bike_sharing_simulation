```python
import requests, zipfile, io

from google.colab import drive
drive.mount('/content/drive')
```

```python
#path
base_path = '/content/drive/MyDrive/complexsystems/'
base_url = "https://s3.amazonaws.com/baywheels-data/"
filename = "-tripdata.csv.zip"
```

```python
for year in range(2018, 2022+1):
  for month in range(1,12+1):

    if year == 2022 and month > 5:
      pass
    
    else:
      str_month = str(month)
      str_year = str(year)
      if month < 10:
        str_month = '0' + str_month

      try:
        url = base_url + str_year +  str_month + '-baywheels' + filename
        r = requests.get(url)
        z = zipfile.ZipFile(io.BytesIO(r.content))
        z.extractall(base_path+"data/raw/")

      except:
        url = base_url + str_year +  str_month + '-fordgobike' + filename
        r = requests.get(url)
        z = zipfile.ZipFile(io.BytesIO(r.content))
        z.extractall(base_path+"data/raw/")
      
      print(url)
```
